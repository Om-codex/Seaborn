## 4.Categorical Categorical Plots
### a.Heatmaps
### b.Clustermaps

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
plt.style.use('fivethirtyeight')
data = sns.load_dataset('flights')
data.head()
```

# Heatmaps
## => A heat map is a graphical representation of data where the individual values contained in a matrix are represented as colors.

```python
pt = data.pivot_table(index='year',columns='month',values='passengers',aggfunc='sum')
```

```python
sns.heatmap(pt,cbar=False,linewidth=0.5,annot=True,fmt='d',cmap='summer')
```

# Clustermap
## => It is basically a heatmap but the only difference is it's clustered , it clusters the rows and columns on the basis of similarities

```python
sns.clustermap(pt,z_score=0,annot=True,col_cluster=False,metric='correlation')
```
