## 3.Categorical Estimate Plots
### a.Pointplot
### b.Barplot
### c.Countplot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
plt.style.use('fivethirtyeight')
data = sns.load_dataset('tips')
data.head()
```

# Barplot 

```python
sns.catplot(x='smoker', y = 'total_bill',palette='Greens',kind='bar',data = data)
```

```python
sns.catplot(x='smoker', y = 'total_bill',kind='bar',hue='sex',data = data)
```

# Countplot
## => Countplot is a special case of Barplot where we do not take any numerical value instead of that we take categorical values and it gives the count of that category

```python
sns.catplot(x='sex',kind='count',data= data,hue='smoker')
```

```python
sns.catplot(x='sex',kind='count',palette='Greens',hue='smoker',data = data)
```
