# Joinplot

## =>Jointplot creates a multi-panel figure that shows both the bivariate (or joint) relationship between two variables along with the univariable (or marginal) distribution of each on separate axes.

# We can use JointPLots in the following four manners:
## 1. Scatter Plots
## 2. Hexbin Plots
## 3. Kdeplots
## 4. Regplots

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.style.use('fivethirtyeight')
data = sns.load_dataset('tips')
data.head()
```

```python
sns.jointplot(x = 'total_bill',y='tip',data=data) 
# A jointplot gives the bivariate as well as univariate plot analysis for each category
```

```python
sns.jointplot(x='total_bill',y='tip',kind='hex',data = data)  #for specific plot specify that specific plot in kind 
```

```python
sns.jointplot(x='total_bill',y='tip',kind='kde',data = data)
```

```python
sns.jointplot(x='total_bill',y='tip',kind='reg',data = data)
```
