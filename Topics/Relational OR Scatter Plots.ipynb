# Relational/Statistical Plots
## Relational plots are used to perform multivariate analysis
## 1.Scatter Plots => kind = 'scatter'
## 2.Line Plots  => kind = 'line'

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.style.use('fivethirtyeight')
```

```python
data = sns.load_dataset('tips')
```

```python
data.head()
```

# #Plot a Scattter Plot between total_bill and tip

```python
sns.relplot(x = 'total_bill',y = 'tip',kind = 'scatter',data = data)
```

# The HUE parameter

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',data = data)        # USE either a relplot or a scatterplot 
sns.scatterplot(x='total_bill',y='tip',hue='smoker',data = data)
```

# The STYLE parameter

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',style='sex',data = data)
```

# The SIZE parameter

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',style='sex',size='size',data = data,sizes=(15,200))
```
