## 2.Categorical Distribution Plots
### a.Boxplot
### b.Violinplot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
plt.style.use('fivethirtyeight')
data = sns.load_dataset('tips')
data.head()
```

# Boxplots

```python
#univariate analysis

sns.boxplot(x=data['total_bill'])
```

```python
sns.catplot(x='day',y='total_bill',kind ='box',data =data)
```

```python
sns.catplot(x='day',y='total_bill',kind ='box',hue='sex',data =data)
```

```python
print('\n')
```

# Violinplots

```python
sns.violinplot(x =data['total_bill'])
```

```python
sns.catplot(x='day',y='total_bill',kind='violin',data = data)
```

```python
sns.catplot(x='day',y='total_bill',kind='violin',hue='sex',split=True,data = data)
```
